```python
class sort:
    def selection_sort(self, qus):
        for i in range(len(qus) - 1):
            m_val = i
            for j in range(i+1, len(qus)):
                if qus[j] < qus[m_val]:
                    m_val = j

            if qus[i] != qus[m_val]:
                qus[i], qus[m_val] = qus[m_val], qus[i]

        return qus

    def selection_sort_using_min_function(self, qus):
        for i in range(len(qus) - 1):
            m_val = min(qus[i:])
            m_ind = qus.index(m_val,i) #this i is used to generate the starting position of the index searching
            if qus[i] != qus[m_ind]:
                qus[i], qus[m_ind] = qus[m_ind], qus[i]
        return qus

    def bubble_sort(self, qus):
        for i in range(len(qus)-1,0,-1):
            for j in range(i):
                if qus[j]>qus[j + 1]:
                    qus[j], qus[j + 1] = qus[j + 1], qus[j]
                    print(qus)

                else:
                    print(qus)
            print()

    def quick_sort(self, qus):
        less = []
        equal =[]
        grater = []

        if len(qus) > 1: #check the length of the array
            pivot = qus[0]
            for i in qus:
                if i == pivot:
                    equal.append(i)

                elif i > pivot:
                    grater.append(i)

                elif i < pivot:
                    less.append(i)

            return self.quick_sort(less) + equal + self.quick_sort(grater) #its recursion method
        else:
            return qus #if have only one array we need to return that value



    def merge_sort(self,qus):
        if len(qus) > 1:
            mid = len(qus)//2
            left = qus[:mid]
            right = qus[mid:]
            self.merge_sort(left)
            self.merge_sort(right)
            i=0
            j=0
            k=0
            while i < len(left) and j< len(right):
                if left[i] < right[j]:
                    qus[k] = left[i]
                    k +=1
                    i += 1

                else:
                    qus[k] = right[j]
                    k += 1
                    j += 1

            while i < len(left):
                qus[k] = left[i]
                i +=1
                k +=1

            while j < len(right):
                qus[k] = right[j]
                j +=1
                k +=1

        return qus

    def insertion_sort(self, qus):
        for i in range(1,len(qus)):
            current_element = qus[i]
            pos = i

            while current_element < qus[pos-1] and pos > 0:
                qus[pos],qus[pos-1] = qus[pos-1],qus[pos]
                pos -= 1
            qus[pos] = current_element


        return  qus



import numpy
qus1 = list(numpy.random.randint(0,50,6))
qus2 = list(numpy.random.randint(0,50,6))
qus3 = list(numpy.random.randint(0,50,6))
qus4 = list(numpy.random.randint(0,50,6))
qus5 = list(numpy.random.randint(0,50,6))
qus6 = list(numpy.random.randint(0,50,6))
ans =sort()
print("selection sort")
print(qus1)
print(ans.selection_sort(qus1))
print("---------------------------------")
print("selection sort using minimum function ")
print(qus2)
print(ans.selection_sort_using_min_function(qus2))
print("---------------------------------")
print("bubble sort")
print(qus3)
ans.bubble_sort(qus3)
print("---------------------------------")
print("quick sort")
print(qus4)
print(ans.quick_sort(qus4))
print("---------------------------------")
print("merge sort")
print(qus5)
print(ans.merge_sort(qus5))
print("---------------------------------")
print("insertion sort")
print(qus6)
print(ans.insertion_sort(qus6))
```
